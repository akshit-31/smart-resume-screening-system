```python
import sys
sys.path.insert(0, '..')

from app.services.preprocessor import preprocess_text
from app.services.tfidf_model import compute_tfidf_similarity
from app.services.ranker import rank_candidates
```

```python
jd = """
We are looking for a Python backend developer with experience in Flask, REST APIs,
PostgreSQL, and machine learning using scikit-learn. Familiarity with Docker and AWS is a plus.
"""

resume1 = """
John Doe — Python Developer
Skills: Python, Flask, REST API, PostgreSQL, scikit-learn, Docker
Experience: 3 years building ML-powered web apps.
"""

resume2 = """
Jane Smith — Frontend Developer
Skills: HTML, CSS, JavaScript, React, Figma
Experience: 2 years in UI/UX design.
"""

jd_clean = preprocess_text(jd)
r1_clean = preprocess_text(resume1)
r2_clean = preprocess_text(resume2)

score1 = compute_tfidf_similarity(jd_clean, r1_clean)
score2 = compute_tfidf_similarity(jd_clean, r2_clean)

print(f'Resume 1 (Python Dev) TF-IDF Score: {score1:.4f}')
print(f'Resume 2 (Frontend Dev) TF-IDF Score: {score2:.4f}')
```
